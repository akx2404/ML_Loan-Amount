```python
df['Dependents'].fillna(1,inplace=True)
```

```python
df['LoanAmount'].fillna(df.LoanAmount.mean(),inplace=True)
df.info()
```

```python
Value_Mapping = {"Yes" : 1, 'No' : 0}
df['Married_Section'] = df['Married'].map(Value_Mapping)
df.head(5)
```

```python
Value_Mapping = {"Male" : 1, 'Female' : 0}
df['Gender_Section'] = df['Gender'].map(Value_Mapping)
df.head(5)
```

```python
Value_Mapping = {"Graduate" : 1, 'Not Graduate' : 0}
df['Edu_Section'] = df['Education'].map(Value_Mapping)
df.head(5)
```

```python
df['Married_Section'].fillna(df.Married_Section.mean(),inplace=True)
df['Gender_Section'].fillna(df.Gender_Section.mean(),inplace=True)
df['Edu_Section'].fillna(df.Edu_Section.mean(),inplace=True)
df['Credit_History'].fillna(df.Credit_History.mean(),inplace=True)
df['Loan_Amount_Term'].fillna(df.Loan_Amount_Term.mean(),inplace=True)

df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df["Property_Section"] = lb_make.fit_transform(df["Property_Area"])
df.head(15)
```

```python
sns.FacetGrid(df,hue="Gender_Section",size=5)\
.map(plt.scatter, "Loan_Status","LoanAmount")\
.add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue="Property_Section",size=5)\
.map(plt.scatter, "ApplicantIncome","CoapplicantIncome")\
.add_legend()
plt.show()
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
Value_Mapping = {"Yes" : 1, 'No' : 0}
df['Employed_Section'] = df['Self_Employed'].map(Value_Mapping)
df['Employed_Section'].fillna(df.Edu_Section.mean(),inplace=True)
```

```python
Value_Mapping = {"Y" : 1, 'N' : 0}
df['Loan_Section'] = df['Loan_Status'].map(Value_Mapping)
df['Loan_Section'].fillna(df.Loan_Section.mean(),inplace=True)
```

```python
X = df[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Married_Section','Gender_Section','Edu_Section','Employed_Section','Property_Section']].values
y = df[["Loan_Section"]].values
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y,test_size=0.2)
```

```python
clf = LinearRegression()
clf.fit(xtrain, ytrain)
```

```python
clf.predict(xtest)
```

```python
clf.score(xtest, ytest)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
clr = LogisticRegression()
```

```python
clr = LogisticRegression()
clr.fit(xtrain, ytrain)
```

```python
clr.score(xtest, ytest)
```

```python
clr.score(xtrain, ytrain)
```
